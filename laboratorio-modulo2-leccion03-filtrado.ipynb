#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
# Filtrar los títulos lanzados en 2021 y que son del tipo 'Movie'
movies_2021 = df_merged[(df_merged["release_year"] == 2021) & (df_merged["type"] == "Movie")]["title"]

# Convertir a lista para mostrar los títulos
movies_2021_list = movies_2021.tolist()
movies_2021_list
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df_pelis_90_min = df_merged.query("Runtime == 90").filter(items= ["title", "Runtime"])
df_pelis_90_min.head()
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
# Filtrar los títulos del país 'India' lanzados después de 2018
df_titulos_india_post_2018 = df_merged[(df_merged["country"] == "India") & (df_merged["release_year"] > 2018)]

# Seleccionar solo las columnas 'title' y 'release_year'
df_titulos_india_post_2018 = df_titulos_india_post_2018[["title", "release_year"]]

# Mostrar los resultados
df_titulos_india_post_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
df_titles_tv14_r = df_merged[df_merged["rating"].isin(["TV-14", "R"])][["title", "rating"]]
df_titles_tv14_r
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
# Filtrar los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'
df_titles_by_nolan_scorsese = df_merged[df_merged["director"].isin(["Christopher Nolan", "Martin Scorsese"])][["title", "director"]]

# Contar cuántos títulos ha dirigido cada director
df_directors_count = df_titles_by_nolan_scorsese["director"].value_counts()

# Mostrar resultados
df_titles_by_nolan_scorsese, df_directors_count
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
# Primer código: Filtrar títulos que pertenecen a los países seleccionados
titles_from_selected_countries = df_merged[df_merged["country"].isin(["United States", "United Kingdom", "Canada"])]
# Mostrar un resumen de los títulos filtrados
titles_from_selected_countries.head()
```

```python
# Asegurarse de que la columna 'Language' sea numérica
titles_from_selected_countries["Language"] = pd.to_numeric(titles_from_selected_countries["Language"], errors='coerce')

# Filtrar títulos con más de un idioma
movies_with_multiple_languages = titles_from_selected_countries[titles_from_selected_countries["Language"] > 1]

# Contar películas con más de un lenguaje por país
movies_with_multiple_languages_count = movies_with_multiple_languages["country"].value_counts()

# Mostrar resultados
movies_with_multiple_languages, movies_with_multiple_languages_count
```

No se encontraron películas con más de un idioma listado en la columna Language. Esto podría deberse a que:

- La mayoría de las películas tienen solo un idioma listado.
- Algunos valores de la columna Language podrían estar vacíos o mal formateados (valores nulos)
